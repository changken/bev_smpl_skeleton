```python
import numpy as np
```

```python
which_camera = 'hd_00_05'
```

```python
# load hd_00_00's joints
results = np.load('D:\\bev_smpl_skeleton\\' + which_camera + '\\video_results.npz',allow_pickle=True)['results'][()]
```

```python
loaded_length = len(results)
```

```python
motion_file= 'D:\\bev_smpl_skeleton\\' + which_camera + '\\' + which_camera + '_HM36Joint.txt'

f = open(motion_file, "a")

for key_name in results:
    print(results[key_name]['joints'][0][54:].shape)
    # 取得human3.6m的17個關鍵點
    joint_points = results[key_name]['joints'][0][54:]
    print(joint_points)
    line = ",".join([str(v) for v in joint_points.flatten().tolist()])
    f.write(line + "\n")    
    print(line)

f.close()
```

```python
# 再次檢查筆數
f2 = open(motion_file, "r")
assert len(f2.readlines()) == loaded_length
f2.close()
```
